```python
import pandas as pd
from matplotlib import pylab as plt
import seaborn as sns
from scipy import stats
from statsmodels.graphics.tsaplots import plot_acf
```

```python
orig_data = pd.read_csv('train_data/ts-data/ts-all.csv',header=0)
```

```python
orig_data.head()
```

```python
sns.pairplot(orig_data[['purchases','unique_invoices','unique_streams','total_views','revenue']])
```

```python
fig = plt.figure(figsize=(14,5))
ax = fig.add_subplot(111)
sns.lineplot(data=orig_data.iloc[::10],x='date',y='revenue',ax=ax)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(111)
plot_acf(orig_data.revenue.values,ax=ax)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(111)
shift = 1
order = 1
temp_data = orig_data.revenue.copy()
for _ in range(order):
    temp_data = (temp_data - temp_data.shift(shift))[shift:]
plot_acf(temp_data,ax=ax)
```
